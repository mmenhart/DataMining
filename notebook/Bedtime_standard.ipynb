```python
def remove_AmPm(series):
    new_series = []
    for line in series:
        
        new1 = line.replace('m', '')
        new2 = new1.replace('p', '')
        new3 = new2.replace('a', '')
        new4 = new3.replace('A', '')
        new5 = new4.replace('M', '')
        new6 = new5.replace('P', '')
        new7 = new6.replace(' ', '')
        new8 = new7.replace('.', ':')
        new_series.append(new8)
        
    print(new_series)
    return new_series
```

```python
# Try to get rid of any am/pm that might be present
jesus = remove_AmPm(odi_df.bedtime)
```

```python
def add_zeros(series):
    new_series = []
    
    justone = r'\d{1,2}(?!(:|\d))'

    
    for line in series:
        m = match(justone, line)
        
        if m is not None:
            line+=':00'
            
        new_series.append(line)
        
    print(new_series)   
    return new_series
            
            
        
```

```python
def nochar(series):
    new_series = []
    
    right = r'\d{2}:\d{2}'
    extra = r'\d*:{1,}'
    
    for line in series:
        r = match(right, line)
        e = match(extra, line)
        
        if r is not None:
            new_series.append(line)
            
        elif e is not None:
            newval = line.replace(':::', ':00')
            new_series.append(newval)
        else:
            new_series.append('NA')
    print(new_series)
    return new_series
                
```

```python
def decimaltime(series):
    new_series = []
    
    for time in series:
        
        
        if time == 'NA':
            pass
        
        else:
            minute = time[-2:]
            if minute != '00':
                new_minute = 60/int(minute)
                new_time = time[:2]
                new_time = new_time+'.'+str(new_minute)
                new_series.append(new_time)

            else:
                new_time = time[:2]
                new_time = new_time+'.'+str(00)
                new_series.append(new_time)
        
        
    print(new_series)    
    return new_series

decimaltime(donzo)
```
