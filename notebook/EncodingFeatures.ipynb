```python
# Import required libraries
import pandas as pd
import numpy as np  

# import csv dataset as pandas dataframe
df = pd.read_csv("../data/standard.csv", sep=',')
```

```python
df1 = df.iloc[:,1:14]
df1.head()
```

```python
to_encode = df1.iloc[:, 0:10]
to_encode.head()
```

```python
encoded = pd.get_dummies(to_encode, prefix=['goodDay1', 'goodDay2', 'program', 'MLcourse', 'IRcourse', 'StatsCourse', 'DBcourse', 'choco_op', 'gender', 'stand'])
encoded.head()
```

```python
encoded['bedtime'] = df1.bedtime.values
```

```python
encoded['neighbors'] = df1.neighbors.values
```

```python
encoded['randomNum'] = df1.randomNum.values
```

```python
encoded.head()
```

```python
encoded.to_csv('../data/encoded.csv')
```
